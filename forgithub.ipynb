```python
# 11. 인덱싱을 통하여 캐릭들의 심볼별 포스 추출
# 11. Cleaning up force data
symbol1 = []
for i in range(len(test1)):
    symbol2 = []
    if 'symbol' in test1[i]:
        for j in range(len(test1[i]['symbol'])):
            symbol2.append(int(test1[i]['symbol'][j]['symbol_force']))
    symbol1.append(symbol2)
symbol1
```

```python
# 12. 캐릭별 총 아케인, 어센틱포스 + 그랜드 어센틱포스 분류 및 데이터에 추가
# 12. Final touches with force data
arcane = []
authentic = []
grand_authentic = []
for i in range(len(symbol1)):
    arcane.append(sum(symbol1[i][:6]))
    authentic.append(sum(symbol1[i][6:]))
    if authentic[i] > 660:
        grand_authentic.append(sum(symbol1[i][12:]))
    else:
        grand_authentic.append(0)

data['arcane'] = arcane
data['authentic'] = authentic
data['grand_authentic'] = grand_authentic
data
```

```python
# 16. 원 데이터에 6차 강화 레벨 붙이기
# 16. Add levels of hexa cores to the original data

data['mastery1_level'] = mastery1_level
data['mastery2_level'] = mastery2_level
data['mastery3_level'] = mastery3_level
data['mastery4_level'] = mastery4_level
data['v1_core_level'] = v1_core_level
data['v2_core_level'] = v2_core_level
data['v3_core_level'] = v3_core_level
data['v4_core_level'] = v4_core_level
data['skill_core_level'] = skill_core_level
data['shared_core_level'] = shared_core_level
```
